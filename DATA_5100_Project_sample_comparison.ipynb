```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

stroke = pd.read_csv('https://raw.githubusercontent.com/kewal97/Stroke-Prediction-Project/main/healthcare-dataset-stroke-data.csv?token=GHSAT0AAAAAAB2DOJ7AKVJ7DYC254JV4DUIY3BWN6Q')
```

```python
def BMI_indicator(x):
  if x < 18.5:
    return 'Underweight'
  elif x <25:
    return 'Healthy'
  elif x< 30:
    return 'Overweight'
  elif x < 35:
    return 'Obese Class 1'
  elif x < 40:
    return 'Obese Class 2'
  else:
    return 'Severely Obese'
```

```python
stroke['BMI_indicator']= stroke['bmi'].apply(BMI_indicator)
```

```python
no_stroke = stroke.loc[stroke["stroke"]==0]
samp_no_stroke = no_stroke.sample(n=1417)
```
